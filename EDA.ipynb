## 1. Importing the Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import eda_helper_functions as eda
from sklearn.ensemble import IsolationForest
```

## 2. Reading the Training Data

```python
file_path = r"data/train.csv"
train = pd.read_csv(file_path)
train
```

```python
train.isnull().sum()
```

```python
train.dropna(inplace=True)
```

```python
train.isnull().sum()
```

```python
train = train.assign(**{
    col : pd.to_datetime(train.loc[:,col], dayfirst=True, format="mixed")
    for col in ["date_of_journey", "dep_time", "arrival_time"]
})
```

## 8.2 date_of_journey

```python
train.date_of_journey
```

```python
eda.dt_univar_plots(
    data = train,
    var = 'date_of_journey',
    target = "price"
)
```

- The price of flights is lowest at the month of April and it peaks during the month of May & June.

```python
(
    train
    .groupby(pd.Grouper(key='date_of_journey', freq='M'))
    .price.mean()
    .to_frame()
    .set_axis(["March","April","May","June"], axis=0)
    .rename_axis(index="month")
    .plot(
        kind='bar',
        figsize=(8,4),
        edgecolor='black'
    )
)

ax = plt.gca()
eda.rotate_xlabels(ax)
```

```python
(
    train
    .groupby(pd.Grouper(key='date_of_journey', freq='W'))
    .price.mean()
    .to_frame()
    #.set_axis(["March","April","May","June"], axis=0)
    .rename_axis(index="week")
    .plot(
        kind='bar',
        figsize=(8,4),
        edgecolor='black'
    )
)

ax = plt.gca()
eda.rotate_xlabels(ax)
```

```python
(
    train
    .groupby([pd.Grouper(key='date_of_journey',freq='M'), "source"])
    .price.mean()
    .unstack(fill_value=0)
    .set_axis(["March","April","May","June"], axis=0)
    .rename_axis(index="month")
    .plot(
        kind='bar',
        figsize=(12,4),
        edgecolor='black'
    )
)

ax = plt.gca()

ax.legend(
    loc='upper left',
    bbox_to_anchor = (1,1)
)

eda.rotate_xlabels(ax)
```

## 8.3 dep_time

```python
train.dep_time
```

```python
eda.dt_univar_plots(train, "dep_time", "price")
```

- Most of the flights are booked from 7 am to 12am

```python
(
    train
    .assign(
        dep_time = lambda df_ : (
            np.select(
                [df_.dep_time.dt.hour.between(4,12, inclusive='left'),
                 df_.dep_time.dt.hour.between(12,16, inclusive='left'),
                 df_.dep_time.dt.hour.between(16,200000000000000000, inclusive='left')],
                ["morning", "afternoon", "evening"],
                default = "night"
            )
        )
    )
    .pipe(eda.num_cat_bivar_plots, "price", "dep_time")
)
```
